```
# Import packages
from getpass import getpass
import logging
import os

import flywheel
```

```
# Instantiate a logger
logging.basicConfig(level=logging.INFO, format='%(asctime)s %(levelname)s %(message)s')
log = logging.getLogger('root')
```

# Flywheel API Key and Client



Get a API_KEY. More on this at in the Flywheel SDK doc [here](https://flywheel-io.gitlab.io/product/backend/sdk/branches/master/python/getting_started.html#api-key).

```
API_KEY = getpass('Enter API_KEY here: ')
```

Instantiate the Flywheel API client

```
fw = flywheel.Client(API_KEY or os.environ.get('FW_KEY'))
```

Show Flywheel logging information

```
log.info('You are now logged in as %s to %s', fw.get_current_user()['email'], fw.get_config()['site']['api_url'])
```

# Main script

By having the `exhaustive=True` flag, it allows you to see an "exhaustive" list of what you have access in the system as a site admin (aka all projects in the system)

```
for project in fw.get_all_projects(exhaustive=True):
  # reload is helpful to load the each project
  project=project.reload()
  #replace with local path of deid_profile.yaml
  project.upload_file('/deid_profile.yaml')
```

# Notes:
- If the de-id file already exists in the project, reuploading file with the same name will overwrite the file.
