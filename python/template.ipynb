## Helper functions

You may need a few helper functions. We recommend that you defined them first in the notebook and use [Google Style Python docstrings](https://sphinxcontrib-napoleon.readthedocs.io/en/latest/example_google.html#example-google-style-python-docstrings) to explicit what they are. Here is an example:

```python
def get_project_id(fw, project_label):
    """Return the first project ID matching project_label
    
    Args:
       fw (flywheel.Client): A flywheel client
       project_label (str):  A Project label
       
    Returns:
       (str): Project ID or None if no project found
    """
    project = fw.projects.find_first(f'label={project_label}')
    if project:
        return project.id
    else:
        return None
```
