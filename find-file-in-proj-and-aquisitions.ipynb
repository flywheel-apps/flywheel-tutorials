# Constants

```python
# Source project.label
PROJECT_LABEL = 'myproject'
```

## Get find first project with that label

```python
project = fw.projects.find_first(f'label={PROJECT_LABEL}') 
```

# How to find files that stored on Flywheel Platform

##1. Files can be attached to the project container directly:


```python
for file in project.files:
    print(file.name)
```

##2. Main files can be attached to acquisition containers that are part of the projects.

* First we will need to get the acquisitions under that project


```python
acquisitions = fw.acquisitions.find(f'parents.project={project.id}')
```

* Then we loop through each aquisition and get the files stored under each aquisition

```python
for acquisition in acquisitions:
    for file in acquisition.files:
        print(file.name)
```

```python
for acquisition in acquisitions:
    for key,value in acquisition.items():
        if key == 'files':
            print('Files:')
            for file in value:
                print(f'  {file.name}')
        elif key == 'permissions':
            pass
        else:
            print(f'{key}:{value}')
    print()
```
